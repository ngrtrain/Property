### Import Datset

```python
import pandas as pd
```

```python
df =pd.read_excel("Property Dataset - Clustering.xlsx",sheet_name='Data Rumah')
```

```python
df.drop(0, axis=0, inplace=True)
```

### Check Missing Value

```python
df.isna().sum()
```

### Replace Missing Value

```python
df['Arah Hadap Rumah'] = df['Arah Hadap Rumah'].fillna(df['Arah Hadap Rumah'].mode()[0])
df['Jumlah Kamar Mandi'] = df['Jumlah Kamar Mandi'].fillna(df['Jumlah Kamar Mandi'].mode()[0])
df['Tingkat/Lantai '] = df['Tingkat/Lantai '].fillna(df['Tingkat/Lantai '].mode()[0])
df['Posisi Rumah'] = df['Posisi Rumah'].fillna(df['Posisi Rumah'].mode()[0])
df['Lebar Jalan Depan Rumah (ROW)'] = df['Lebar Jalan Depan Rumah (ROW)'].fillna(df['Lebar Jalan Depan Rumah (ROW)'].mode()[0])
df['Arah Hadap Rumah'] = df['Arah Hadap Rumah'].fillna(df['Arah Hadap Rumah'].mode()[0])
df.isna().sum()
```

### Melabeli Data

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Nama Daerah'] = encoder.fit_transform(df['Nama Daerah'])
df['Area Category'] = encoder.fit_transform(df['Area Category'])
df['Terjual/Belum'] = encoder.fit_transform(df['Terjual/Belum'])
df['Arah Hadap Rumah'] = encoder.fit_transform(df['Arah Hadap Rumah'])
df['Posisi Rumah'] = encoder.fit_transform(df['Posisi Rumah'])
df['Lebar Jalan Depan Rumah (ROW)'] = encoder.fit_transform(df['Lebar Jalan Depan Rumah (ROW)'])
df
```

```python
df.dtypes
```

### Memperbaiki data 

```python
for i in df.columns:
    print('4 98' in df[i].values)
```

```python
df[df['Luas Bangunan (m2)'] == '4 98']
```

```python
df['Luas Bangunan (m2)'] = df['Luas Bangunan (m2)'].replace(['4 98'], 4.98)
```

```python
df[df['Harga Penawaran (dari Owner)'] == ' ']
df['Harga Penawaran (dari Owner)'] = df['Harga Penawaran (dari Owner)'].replace([' '], df['Harga Penawaran (dari Owner)'].mode()[0])
dfcolumns = df.columns
```

### Scalling data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df = sc.fit_transform(df)
```

```python
df = pd.DataFrame(df,columns=dfcolumns)
df
```

### Menentukan jumlah cluster

```python
from sklearn.cluster import KMeans

k_range = range(1,11)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df.iloc[:,:])
    sse.append(km.inertia_)
from kneed import KneeLocator
kl = KneeLocator(range(1,11),sse,curve="convex",direction="decreasing")
kl.elbow
```

```python
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(df.iloc[:,:])
df['Cluster'] = y_pred
df
```

### Meanmpilkan Dendogram

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.title(" Dendrogram")

selected_data = df.iloc[:, :-1]
dendrogram = sch.dendrogram(sch.linkage(selected_data, method='complete'))
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.title(" Dendrogram")

selected_data = df.iloc[:, :-1]
dendrogram = sch.dendrogram(sch.linkage(selected_data, method='ward'))
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.title(" Dendrogram")

selected_data = df.iloc[:, :-1]
dendrogram = sch.dendrogram(sch.linkage(selected_data, method='single'))
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.title(" Dendrogram")

selected_data = df.iloc[:, :-1]
dendrogram = sch.dendrogram(sch.linkage(selected_data, method='average'))
plt.show()
```
